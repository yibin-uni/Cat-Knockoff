# Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
train = pd.read_csv('../data/train.csv')
test = pd.read_csv('../data/test.csv')
x_train = train.drop(['Id','SalePrice'],axis=1)
y_train = train['SalePrice']
x_train
```

## Missing Values

```python
missing = train.isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)
missing.plot.bar(figsize = (8,6), title = "Missing Values")
plt.show()
```

```python
train.describe()
```

## Plot the distributions of some variables

```python
num_cols = list(x_train.columns[x_train.dtypes != 'object'])
print('Numerical variables:')
print(num_cols)
```

```python
cat_cols = list(x_train.columns[x_train.dtypes == 'object'])
print('Categorical variables:')
print(cat_cols)
```

### Distribution of the Response `SalePrice`

```python
import warnings 
warnings.filterwarnings("ignore")

import scipy.stats as st
y = train['SalePrice']
plt.figure(1,figsize=(8,6)); plt.title('Johnson SU')
sns.distplot(y, kde=False, fit=st.johnsonsu)
plt.figure(2,figsize=(8,6)); plt.title('Normal')
sns.distplot(y, kde=False, fit=st.norm)
plt.figure(3,figsize=(8,6)); plt.title('Log Normal')
sns.distplot(y, kde=False, fit=st.lognorm)
```

### Distributions of Numerical Variables

```python
f = pd.melt(train.drop('Id',axis=1), value_vars=num_cols)
g = sns.FacetGrid(f, col="variable",  col_wrap=4, sharex=False, sharey=False)
g = g.map(sns.distplot, "value")
plt.show()
```

### Distributions of the Categorical Variables

```python
# regard missing values as another category named "MISSING"
for c in cat_cols:
    train[c] = train[c].astype('category')
    if train[c].isnull().any():
        train[c] = train[c].cat.add_categories(['MISSING'])
        train[c] = train[c].fillna('MISSING')
# plot the distributions of categorical variables in box plots
def boxplot(x, y, **kwargs):
    sns.boxplot(x=x, y=y)
    x=plt.xticks(rotation=90)
f = pd.melt(train, id_vars=['SalePrice'], value_vars=cat_cols)
g = sns.FacetGrid(f, col="variable",  col_wrap=2, sharex=False, sharey=False, size=5)
g = g.map(boxplot, "value", "SalePrice")
```

## Plot some correlations

Make ordinal encoding for the categorical variables

To keep the missing values intact when making ordinal encoding, we need `sklearn` >= 0.24 

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
enc.fit(train[cat_cols])
ord_cat = pd.DataFrame(enc.transform(train[cat_cols]),columns = train[cat_cols].columns)
train[cat_cols] = ord_cat
train
```

```python
def spearman(frame, features):
    spr = pd.DataFrame()
    spr['feature'] = features
    spr['spearman'] = [frame[f].corr(frame['SalePrice'], 'spearman') for f in features]
    spr = spr.sort_values('spearman')
    plt.figure(figsize=(6, 0.25*len(features)))
    sns.barplot(data=spr, y='feature', x='spearman', orient='h')
    
features = num_cols + cat_cols
spearman(train, features)
```

```python
plt.figure(1,figsize=(12,8))
corr = train[num_cols+['SalePrice']].corr()
sns.heatmap(corr,cmap=plt.cm.coolwarm)
plt.figure(2,figsize=(12,8))
corr = train[cat_cols+['SalePrice']].corr()
sns.heatmap(corr,cmap=plt.cm.coolwarm)
plt.figure(3,figsize=(12,8))
corr = pd.DataFrame(np.zeros([len(num_cols)+1, len(cat_cols)+1]), index=num_cols+['SalePrice'], columns=cat_cols+['SalePrice'])
for q1 in num_cols+['SalePrice']:
    for q2 in cat_cols+['SalePrice']:
        corr.loc[q1, q2] = train[q1].corr(train[q2])
sns.heatmap(corr,cmap=plt.cm.coolwarm)
```
